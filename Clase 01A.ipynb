## Campo de direcciones

Corresponde a un bosquejo con pequeños segmentos de recta trazados en un sistema de coordenadas cartesianas $xy$ (o simplemente plano $xy$), donde se muestra el comportamiento de la pendiente de la curva solución a la EDO $y'=f(x,y)$.

Para generar dicho bosquejo en Python, podemos utilizar las funciones anteriormente mencionadas, en este sentido, en vez de graficar "pequeños segmentos" en la práctica graficamos vectores unitarios en las direcciones correspondientes.


## Ejemplo 2

Para iniciar nuestro estudio consideremos la EDO vista en clase teórica
$$ y'= 2y $$
y generemos el campo de direcciones asociado a la ecuación.

```python
X,Y=np.meshgrid(np.linspace(-4,4,9), np.linspace(-4,4,9))

U = 1.0
V = 2*Y

N = np.sqrt(U ** 2 + V ** 2) #Normalizamos los vectores
U = U / N
V = V / N

plt.quiver(X, Y, U, V) #graficamos los vectores con origen en (x,y) y término (u,v)
plt.xlim([-5, 5]) #limites en eje x
plt.ylim([-5, 5]) #limites en eje y

plt.xlabel('x') #etiqueta eje x
plt.ylabel('y') #etiqueta eje y
plt.grid()# para establecer red en la región 
```

## EDO autónoma

Una ecuación de la forma $y'=f(y)$, es decir, una ecuación diferencial de primer orden en donde $f$ depende solo de la variable dependiente se denomina ecuación autónoma.

## Ejemplo 4

a) Identificar cuál de las siguientes EDO son autónomas.
  *  $y'=3x-y^2$
  *  $y'= (y-3)\ln(y^2+1)$
  *  $ (y-1)(3-y)dx-dy=0 $
 
b) Graficar los campos de direcciones asociados SOLO a las EDOs autónomas del item anterior.

```python
'''
a) La opciones 2 y 3 don autonomas, la primera se descarga pues f es en función de x e y
'''
```

```python
# Campo de direcciones de y' = (y-3)*ln(y**2 -1)
import numpy as np
import matplotlib.pyplot as plt


# EDO AUTONOMA 2
X,Y = np.meshgrid(np.linspace(-5, 5, 12), (np.linspace(-5, 5, 12)))

U = 1.0
V = (Y-3)* np.log(Y**2+1)

N = np.sqrt(U**2 + V**2)

U = U / N
V = V / N

plt.quiver(X, Y, U, V)
plt.xlim([-5, 5])
plt.ylim([-5, 5])

plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.title("Campo direccional autonoma 2")

plt.show()

# EDO AUTONOMA 3
X, Y = np.meshgrid(np.linspace(-5, 5, 12), np.linspace(-5, 5, 12))
U = 1.0
V = (Y-1)*(3-Y)

N = np.sqrt(U**2 + V**2)
U = U/N
V = V/N

plt.quiver(X, Y, U, V)
plt.xlim([-5, 5])
plt.ylim([-5, 5])

plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.title("Campo direccional autonoma 3")

```

## Puntos de equilibrio

Dada una EDO autónoma $y'=f(y)$, si $f(y_0) = 0$ diremos que la recta $y = y_0$ es una solución de equilibrio de la EDO autónoma, y al valor $y_0$ lo denominamos punto de equilibrio.

### Clasificación de puntos de equilibrio

* Un punto de equilibrio $y_0$ se dirá sumidero o atractor si cualquier solución suficientemente cerca a $y = y_0$ tiende asintóticamente a $y_0$ a medida de que $x$ aumenta.

* Un punto de equilibrio $y_0$ se dirá fuente o repulsor si cualquier solución suficientemente cerca a $y = y_0$ tiende asintóticamente a $y_0$ a medida de que $x$ decrece.

* Un punto de equilibrio $y_0$ se dirá nodo si este no es fuente ni sumidero.




### Ejemplo 6:  $y'=2-y$

```python
X, Y = np.meshgrid(np.linspace(-5, 5, 20), np.linspace(-5, 5, 20))
U = 1.0
V = 2-Y
N = np.sqrt(U**2 + V**2)
U = U/N
V = V/N
plt.quiver(X, Y, U, V)
plt.grid()
plt.xlim([-5, 5])
plt.ylim([-5, 5])
plt.xlabel('x')
plt.ylabel('y')

# PUNTOS DE EQUILIBRIO, Para que 2-y = 0, y = 2
x = np.linspace(-5, 5, 200)
y = np.linspace(2, 2, 200)
plt.plot(x, y)

# Para y=2 tiene a ese número cuando x aumenta, por lo tanto es punto de equilibrio sumidero o atractor
```

### Ejemplo 7: $y'=y-2$

```python
X,Y = np.meshgrid(np.linspace(-5, 5, 20), np.linspace(-5, 5, 20))
U = 1.0
V = Y-2
N = np.sqrt(U**2 + V**2)
U = U/N
V = V/N
plt.quiver(X,Y,U,V)
plt.grid()
plt.xlim([-5, 5])
plt.ylim([-5, 5])
plt.xlabel('x')
plt.ylabel('y')

# Para que y-2= 0, y=2
x = np.linspace(-5, 5, 200)
y = np.linspace(2, 2, 200)
plt.plot(x, y)

# A medida que x crece se aleja el cambio de y=2, mientras que si se acerca cuando x decrece, por lo tanto es
# punto de equilibrio tipo fuente o repulsor
```

### Ejemplo 8: $y'=(y-1)^2$

```python
X,Y = np.meshgrid(np.linspace(0, 10, 20), np.linspace(0, 10, 20))
U = 1.0
V = (Y-1)**2
N = np.sqrt(U**2 + V**2)
U = U/N
V = V/N
plt.quiver(X,Y,U,V)
plt.grid()
plt.xlim([0, 10])
plt.ylim([0, 10])
plt.xlabel('x')
plt.ylabel('y')

# Para que (y-1)**2 = 0, y = 1
x = np.linspace(0,10, 200)
y = np.linspace(1, 1, 200)
plt.plot(x, y)

# Observando hay puntos que se acerca a 1 a medida que x crece pero otros lo repelen, por lo tanto es nodo
```
